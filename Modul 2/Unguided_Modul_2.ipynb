```python
# Soal 1 Load data (movie classification)

import pandas as pd

df=pd.read_csv('/Movie_classification.csv')
df.info()
```

```python
# Melihat data atas
df.head()
```

```python
# Melihat data bawah
df.tail()
```

```python
# Soal 2 Cek nilai duplikat
df.duplicated()
```

```python
# Soal 3 Menemukan null values buat presentase

null_counts = df.isnull().sum()
null_percentage = (null_counts / len(df)) * 100
print(null_percentage)
```

```python
# Soal 4 Drop missing value berdasarkan baris, kolom, imputasi dengan mean, modus, median

 # Menghapus baris yang nilainya null
df_dropped_rows = df.dropna()
df_dropped_rows.isna ().sum()
```

```python
# Menghapus kolom yang nilainya null
df_dropped_cols = df.dropna(axis=1)
df_dropped_cols.isna ().sum()
```

```python
# imputasi dengan mean
df_imputasi_mean = df.fillna(df.select_dtypes('number').mean())
df_imputasi_mean
```

```python
# imputasi dengan modus
df_imputasi_mode = df.fillna(df.select_dtypes('number').mode())
df_imputasi_mode
```

```python
# imputasi dengan median
df_imputasi_median = df.fillna(df.select_dtypes('number').median())
df_imputasi_median
```

```python
# soal 5 Export data ke csv dan excel
df.to_csv('/data_exported.csv', index=False)
df.to_excel('/data_exported.xlsx', index=False)
```
